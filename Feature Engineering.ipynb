### Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`
- 5% ì´ìƒ -> `A`
- -5 ~ 5% -> `B`
- -5%ì´í•˜ -> `C`
- 10%ì´í•˜ -> `D`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

```
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g_0.csv'

import pandas as pd
import numpy as np 

column_headers = ['ë¶„ê¸°', 'ë§¤ì¶œì•¡', 'ì˜ì—…ì´ìµ', 'ì˜ì—…ì´ìµ(ë°œí‘œê¸°ì¤€)', 'ì„¸ì „ê³„ì†ì‚¬ì—…ì´ìµ',
           'ë‹¹ê¸°ìˆœì´ìµ', 'ë‹¹ê¸°ìˆœì´ìµ(ì§€ë°°)', 'ë‹¹ê¸°ìˆœì´ìµ(ë¹„ì§€ë°°)', 'ìì‚°ì´ê³„', 'ë¶€ì±„ì´ê³„',
           'ìë³¸ì´ê³„', 'ìë³¸ì´ê³„(ì§€ë°°)', 'ìë³¸ì´ê³„(ë¹„ì§€ë°°)', 'ìë³¸ê¸ˆ', 'ì˜ì—…í™œë™í˜„ê¸ˆíë¦„',
           'íˆ¬ìí™œë™í˜„ê¸ˆíë¦„', 'ì¬ë¬´í™œë™í˜„ê¸ˆíë¦„', 'ì˜ì—…ì´ìµë¥ ', 'ìˆœì´ìµë¥ ', 'ROE(%)',
           'ROA(%)', 'ë¶€ì±„ë¹„ìœ¨', 'ìë³¸ìœ ë³´ìœ¨', 'EPS(ì›)', 'PER(ë°°)']
df = pd.read_csv(url, names = column_headers, thousands = ',')
df = df.fillna(0)
df.mean()

df['ë§¤ì¶œì•¡ í‰ê· '] = df['ë§¤ì¶œì•¡'] / 12550.800
df['ë§¤ì¶œì•¡ í‰ê· '] = (df['ë§¤ì¶œì•¡ í‰ê· ']-1)*100

grades = []
for x in df['ë§¤ì¶œì•¡ í‰ê· ']:
    if x >= 10:
        grades.append("S")
    elif x >= 5:
        grades.append("A")
    elif  x >= -5:
        grades.append("B")
    elif x < -5: 
        grades.append("C")
    elif x < -10:
        grades.append("D")
    else:
        grades.append("")

df['ë§¤ì¶œ ì ìˆ˜'] = grades
df

```

```
avg = df['ë§¤ì¶œì•¡'].mean()
print(avg)

#í•´ë‹¹ ë“±ê¸‰ì„ ìœ„í•´ í•„ìš”í•œ ë§¤ì¶œì•¡
print(round(avg*1.1,4),'S')
print(round(avg*1.05,4),'A')
print(round(avg*0.95,4),'B')
print(round(avg*0.9,4),'C')
```
